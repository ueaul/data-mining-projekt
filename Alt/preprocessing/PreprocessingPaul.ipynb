# Preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
df_target = df[["salary_in_usd"]]
df_dropped = df.drop(columns=["salary", "salary_in_usd"])

#work_year ordinal oder oneHot ausprobieren
#company_size ordinal oder oneHot ausprobieren

oneHot_features = ["work_year", "employment_type", "remote_ratio", "job_title", "salary_currency", "employee_residence", "company_location","company_size"]
ordinal_features = ["experience_level"]
numeric_features = []
preprocessor = ColumnTransformer(transformers=[
   ('num', StandardScaler(), numeric_features),
    ('cat', OneHotEncoder(sparse_output=False), oneHot_features),
    ('ord', OrdinalEncoder(categories=[
        ['EN', 'MI', 'SE', 'EX'],
    ]), ordinal_features)])

preprocessor.set_output(transform="pandas")
df_preprocessed = preprocessor.fit_transform(df_dropped)
df_preprocessed
```
